```python
import math as m     # math 모듈을 사용한다.

# 주어진 두 개의 정수와 연산자를 사용하여 계산을 수행한다.
def Calculator(num1, num2, oper):
    if oper == '+':
        return num1 + num2
    elif oper == '-':
        return num1 - num2
    elif oper == '*':
        return num1 * num2
    elif oper == '/':
        if num2 == 0:             # 두 번째 정수가 0인 경우 오류 발생
            raise ZeroDivisionError('나누는 수는 0이 될 수 없습니다.')
        else:
            return num1 / num2
    elif oper == '**':
        return int(m.pow(num1, num2))
    else:                         # 지원하지 않는 연산자가 입력되면 오류 발생
        raise Exception('지원하지 않는 연산자를 입력하였습니다.')

# 사용자가 입력한 값이 정수인지 확인하고 정수가 입력될 때까지 반복해서 입력은 받는다.
def get_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:         # 입력값이 정수가 아닌 경우 오류 발생
            print("잘못된 입력입니다. 정수를 입력해주세요.")


while True:
    try:
        num1 = get_int("첫번째 정수를 입력하세요: ")
        num2 = get_int("두번째 정수를 입력하세요: ")
        oper = input('연산자를 입력하세요: ');
        result = print(Calculator(num1, num2, oper))

    except ValueError as ve:
        print(ve)

    except ZeroDivisionError as ze:
        print(ze)

    except Exception as expt:
        print(expt)

    finally:                        # 계속 계산을 진행할지 확인한다.
        con = input("계속 진행하시겠습니까? (Y/N): ").lower()
        if con != 'y':
            print("계산기를 종료합니다.")
            break
```
